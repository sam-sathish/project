# matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Bar chart

```python
student =['english.grade','math.grade','sciences.grade','language.grade']
percentage=[3.5,3.7,3.1,1]
```

```python
plt.bar(student, percentage,color="blue",edgecolor="black")
```

# Scatter plot

```python
plt.scatter(student, percentage,color="red")
```

# Line plot

```python
plt.plot(student, percentage, linestyle="solid")
```

# Pie chart

```python
plt.pie(percentage, labels=student)
```
